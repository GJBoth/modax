In this notebook we analyze the BIC from the very long run, mostly seeing where the minimum is and relative factors etc

```python
from deepymod.analysis import load_tensorboard

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
df = load_tensorboard('runs/BIC_initial_kdv/')
```

```python
plt.plot(df.BIC)
plt.axvline(df.BIC.idxmin(), linestyle='--', color='k')
```

We can also plot the $\delta$ BIC compared to the minimum:

```python
plt.semilogy(df.BIC - df.BIC.min())
```
